# Ejercicio 2: Matriz TF-IDF

```python
import unicodedata
import pandas as pd
import math

# Función para eliminar tildes
def remove_accents(text):
    return ''.join(c for c in unicodedata.normalize('NFD', text)
                   if unicodedata.category(c) != 'Mn')

# Cargar corpus
corpus = open("01_corpus_turismo.txt", "r", encoding="utf-8")
corpus = corpus.readlines()
corpus = [corpus.strip() for corpus in corpus]

# Construir diccionario sin tildes
dictionary = []

for doc in corpus:
  words = doc.split()
  for word in words:
    word_no_accents = remove_accents(word.lower())
    if word_no_accents not in dictionary:
      dictionary.append(word_no_accents)

# Matriz TF
tf_matrix = []
df_counts = {}

for term in dictionary:
  array = []
  df_term = 0
  for doc in corpus:
    count = 0
    term_found = False
    for word in doc.split():
      word_no_accents = remove_accents(word.lower())
      if word_no_accents == term:
        count += 1
        term_found = True
    if term_found:
      df_term += 1
    array.append(count)
  tf_matrix.append(array)
  df_counts[term] = df_term

# Matriz IDF
N = len(corpus)
idf_values = []
for term in dictionary:
  df_term = df_counts[term]
  idf = math.log(N / (1 + df_term))
  idf_values.append(idf)

# Impresión de la matriz TF
tf_df = pd.DataFrame(tf_matrix, index=dictionary, columns=corpus)
print("Matriz TF:")
display(tf_df)

# IMpresión de la matriz IDF
idf_df = pd.DataFrame(idf_values, index=dictionary, columns=["IDF"])
print("Matriz IDF:")
display(idf_df)

```
